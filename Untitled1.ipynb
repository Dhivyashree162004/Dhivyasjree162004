```python

# Install dependencies
!pip install gradio nltk pandas scikit-learn

# Import necessary libraries
import gradio as gr
import nltk
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK stopwords
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Sample dataset (replace with any dataset)
documents = [
    "Machine learning is a method of data analysis that automates analytical model building.",
    "Artificial intelligence enables machines to learn from experience and perform human-like tasks.",
    "Natural language processing helps computers understand and interpret human language.",
    "Deep learning is a subset of machine learning that uses neural networks.",
    "Information retrieval is the process of obtaining relevant information from large data sources."
]

# Preprocess text: Tokenization & Stopword removal
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Convert to lowercase and tokenize
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return " ".join(filtered_tokens)

preprocessed_docs = [preprocess_text(doc) for doc in documents]

# Convert text to document-term matrix
vectorizer = CountVectorizer()
dt_matrix = vectorizer.fit_transform(preprocessed_docs)

# Train LDA model
lda = LatentDirichletAllocation(n_components=2, random_state=42)  # Extract 2 topics
lda.fit(dt_matrix)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Function to display topic words
def get_topic_words(model, feature_names, n_words=5):
    topics = []
    for topic_idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[:-n_words - 1:-1]]
        topics.append(f"Topic {topic_idx+1}: {', '.join(top_words)}")
    return "\n".join(topics)

# Define topic extraction function
def extract_topics(text):
    processed_text = preprocess_text(text)
    text_vectorized = vectorizer.transform([processed_text])
    topic_distribution = lda.transform(text_vectorized)[0]

    # Get the most dominant topic
    topic_idx = topic_distribution.argmax()
    topics_info = get_topic_words(lda, feature_names)

    return f"Predicted Topic: Topic {topic_idx + 1}\n\nTop Words in Topics:\n{topics_info}"

# Create Gradio interface
interface = gr.Interface(
    fn=extract_topics,
    inputs=gr.Textbox(label="Enter Text"),
    outputs=gr.Textbox(label="Detected Topic"),
    title="Topic Modeling System",
    description="Enter a document, and the system will predict the most relevant topic and show key words from all topics."
)

# Launch Gradio app
interface.launch(share=True)
```
